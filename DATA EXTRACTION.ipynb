**SOME IN BUILT FUNCTIONS**:

--df.axes tell rows in our data.

--df.dtypes tell us data types in our data

--df.empty tells if is there any missing value in our data

--df.ndim tells us dimensioon of our data

--df.shape tells us the shape of our data

--df.isnull().sum() tell us which column has how many missing value

--df.dropna() removes the column which has missing value

--df.size() gives us the size of data rows*column

--df.values() gives us anarray of for df

--df.copy gives us copy of our data frame

--df.sort_values(by='column') sort any column in ascending order


--df['columnname']=df['columnname'].astype(type to be converted into)

--df['columnname']=df['columname'].add(value to be added) add the value to the values of given column

--df=df.filter(items=['column','column'...]) or df[['column','column']] gives us filtered data frame with only mentioned columns

--df.to_dict() save out dataframe to a dictionary.

--df.to_string() saves in string.

--df.columns gives us only column name

--df.columns[index] gives us column of that index

--df.columns.tolist list of column labels

--df.columns.value gives array of columns

--df.rename(columns={'previous_name':'newname',....}) renames our columns name

--df['newcolumnname']=df['condtion column'].where(condition,other=0) gives us a new clumn satisfying the condition
