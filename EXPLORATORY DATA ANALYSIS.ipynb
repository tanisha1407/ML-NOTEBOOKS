***EXPLORATORY DATA ANALYSIS***

There are three parts of EDA:

--UNIVARIATE ANALYSIS:  Analysis on a single independent column.

--BIVARIATE ANALYSIS: Analysis on two columns.

--MULTIVARIATE ANALYSIS: Analysis on more than 2 columns.

There are two types of data numerical and categorical data.

Numerical data is continous data like age,height,weight etc.

Categorical data is discrete data like no of employees etc.

BIVARIATE ANALYSIS:

--scatterplot(numerical column-numerical column)

--sns.scatterplot(x=filename[column],y=filename[column])


```python
tips=pd.read_csv('tips.csv')
tips
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'])
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],data=tips,hue=tips['sex'])
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],data=tips,hue=tips['sex'],style=tips['smoker'])
```
